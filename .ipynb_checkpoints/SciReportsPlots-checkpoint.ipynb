### Degree power estimation.

For this experiment, with 1000 nodes (completed) and 10000 nodes (in progress), we generate networks using the degree power model with power alpha going from 0.0, 0.1, to 2.0, and estimate the power using MLEs (trying values -0.1, 0.09, ..., 2.1) 

```python
root = "estimatepower/likelihoods/like-1000-"

params = np.linspace(0.0,2.0,21)
guesses = np.zeros((10,len(params)),dtype=float)

for j, p in enumerate(params):
    file = root+str(round(p,1))+".dat"
    with open(file,'r') as f:
        for i, line in enumerate(f):
            guesses[i,j] = float(line.strip())
            
means = np.mean(guesses,axis=0)
sds = np.std(guesses,axis=0)

a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

plt.xlabel("Parameter $\\alpha$", fontsize=30)
plt.ylabel("MLE $\\hat{\\alpha}$", fontsize=30)
plt.tick_params(labelsize=20)

plt.plot(params, means, linewidth=3, marker = "o")
plt.fill_between(params, means - 1.96*sds, means + 1.96*sds, alpha = 0.3)

plt.plot(params,params, linestyle="--", color="black")

plt.show()
```
